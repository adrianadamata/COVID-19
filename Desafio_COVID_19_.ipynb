# Porcentagem de valores ausentes
Vamos avaliar agora qual a porcentagem de dados ausentes em cada variável. Os valores ausentes precisam ser preenchidos ou o item removido do dataframe, para que possa ser implementado algum método de Machine Learning. A partir das análises podemos concluir que:

* 

````
# porcentagem de valores ausentes em cada variável
df.append((df.isnull().sum() / df.shape[0]),ignore_index=True)
````

````
# remover as colunas *ausentes* em um novo DataFrame

#tresh refer-se aos valores que não são NaN, o mínimo necessário para a coluna sobreviver. 
#Nesse exemplo, considerei que precise de pelo menos 1 valor não ausente para a coluna ser mantida:
df_clean = df.dropna(axis=1, thresh =1)
df_results = df_clean['SARS-Cov-2 exam result'].str.replace('negative', '0').str.replace('positive', '1').astype("int64")
display(df_clean)
````

Observa-se que 5 colunas foram eliminadas pois não tinham nenhuma informação sobre resultados.

````
# criar uma matriz de correlação
corr = df_clean.corr(df_results)
# mostrar a matriz de correlação
display(corr)
````
